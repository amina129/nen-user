```python
def solve_and_show(grid):
    rows = len(grid)
    cols = len(grid[0])

    # Print grid
    print("\n╔════════════════╗")
    print("║  NEN GRID      ║")
    print("╚════════════════╝\n")

    for i in range(rows):
        line = "  "
        for j in range(cols):
            line += f"[{grid[i][j]}]"
        print(line)

    # Find path
    path = []
    r = c = 0
    path.append((r, c))

    while r != rows-1 or c != cols-1:
        current = grid[r][c]

        # Check right
        if c+1 < cols and grid[r][c+1] > current:
            c += 1
            path.append((r, c))
        # Check down
        elif r+1 < rows and grid[r+1][c] > current:
            r += 1
            path.append((r, c))
        else:
            # Backtrack
            if len(path) > 1:
                path.pop()
                r, c = path[-1]
            else:
                break

    # Show result
    print("\n" + "-"*20)
    if (r, c) == (rows-1, cols-1):
        print("✅ PATH FOUND!")
        print("\nPath coordinates:")
        for step in path:
            print(f"  {step} = {grid[step[0]][step[1]]}")
    else:
        print("❌ NO PATH")

    print("\n" + "="*20)

# Test it
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

solve_and_show(grid)
```
